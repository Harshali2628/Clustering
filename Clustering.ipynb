#THEORETICAL QUESTIONS

1. What is unsupervised learning in the context of machine learning?
    - Unsupervised learning is a type of machine learning where algorithms learn patterns from unlabeled data, i.e., there are no given target outputs or class labels. The goal is typically to discover hidden structure, such as clusters or lower‑dimensional representations, by relying only on the inherent similarities and relationships among data points.

2. How does K-Means clustering algorithm work?
    - K‑Means is an iterative, centroid‑based clustering algorithm that partitions data into K clusters. The basic steps are:

    - Choose the number of clusters K and initialize K centroids (often randomly).

    - Assign each data point to the nearest centroid (usually using Euclidean distance), forming K clusters.

    - Recompute each centroid as the mean of all points assigned to that cluster.

    - Repeat the assign–recompute steps until centroids stabilize or a maximum number of iterations is reached.

    - The objective is to minimize the within‑cluster sum of squared distances (inertia).

3. Explain the concept of a dendrogram in hierarchical clustering.
    - A dendrogram is a tree‑like diagram that visualizes the sequence of merges or splits in hierarchical clustering. On the x‑axis, it shows individual data points or clusters, and on the y‑axis, it shows the distance (or dissimilarity) at which clusters are merged. Cutting the dendrogram at a chosen height gives a particular clustering: all branches below the cut form the final groups, making it easy to see the nested structure of clusters at different levels of granularity.

4. What is the main difference between K-Means and Hierarchical Clustering?
    - K‑Means is a partitional algorithm: it directly partitions the data into K clusters, and you must specify K in advance. It is iterative and can reassign points between clusters during optimization.

    - Hierarchical clustering builds a hierarchy of clusters (either bottom‑up agglomerative or top‑down divisive) and produces a dendrogram; it does not require a pre‑specified K and usually does not revise previous merge/split decisions.

    - In practice, K‑Means is typically faster and more scalable, while hierarchical clustering gives more interpretable, multi‑level cluster structures.

5. What are the advantages of DBSCAN over K-Means?
    - DBSCAN (Density‑Based Spatial Clustering of Applications with Noise) has several key advantages:

    - It does not require the number of clusters to be specified in advance; clusters emerge from density parameters (eps and minPts).

    - It can find clusters of arbitrary shape (e.g., non‑spherical, curved structures), whereas K‑Means tends to favor roughly spherical clusters due to its centroid‑and‑variance objective.

6. When would you use Silhouette Score in clustering?
    - The Silhouette Score is used to evaluate how well the clustering structure fits the data, without requiring ground‑truth labels. It measures, for each point, how similar it is to its own cluster compared to other clusters; scores range from −1 to 1, where higher values indicate better defined, well‑separated clusters. It is commonly used to:

    - Compare different clustering algorithms or parameter settings (e.g., different K in K‑Means).

    - Help choose an appropriate number of clusters by selecting the K that yields a higher average Silhouette Score.

7. What are the limitations of Hierarchical Clustering?
    - Limitations of hierarchical clustering

    - Computationally expensive: naive implementations have at least O(n^2) time and memory complexity, so they do not scale well to very large datasets.

    - Once clusters are merged (in agglomerative) or split (in divisive), those decisions are not revised, which can propagate early mistakes through the hierarchy.

8. Why is feature scaling important in clustering algorithms like K-Means?
    - K‑Means relies on distance computations (typically Euclidean) to assign points to the nearest centroid. If features are on very different scales (e.g., income in thousands vs. age in years), the larger‑scale feature dominates the distance, effectively overpowering other features. Feature scaling (e.g., standardization to zero mean and unit variance, or min‑max normalization) ensures all features contribute more fairly, leading to more meaningful cluster assignments and centroids.

9. How does DBSCAN identify noise points?
    - DBSCAN uses two parameters: eps (neighborhood radius) and minPts (minimum number of points to form a dense region). For each point:

    - If it has at least minPts neighbors within distance eps, it is a core point.

    - Points reachable from a core point (within chains of eps‑neighbors) form a cluster.

    - Any point that is neither a core point nor reachable from any core point is labeled as noise (or outlier).

    - Thus, noise points are those in low‑density regions that do not belong to any dense cluster.

10. Define inertia in the context of K-Means.
    - Inertia in K‑Means is the within‑cluster sum of squared distances between each data point and the centroid of its assigned cluster. Formally, for each cluster j with centroid cj and points xi in that cluster, inertia is

    - Inertia= K∑j=1 ∑xi∈Cj ∥xi − cj∥^2
    - Lower inertia indicates that points are closer to their centroids, meaning more compact clusters, but inertia alone should not be used blindly for model selection because it always decreases as K increases.

11. What is the elbow method in K-Means clustering?
    - The elbow method is a heuristic to choose a good number of clusters K by plotting K versus a clustering cost (usually inertia, i.e., within‑cluster sum of squared distances). As K increases, inertia decreases, and the “elbow” is the point where the rate of decrease sharply slows, suggesting a reasonable trade‑off between model complexity and cluster compactness.

12. Describe the concept of "density" in DBSCAN.
    - In DBSCAN, density refers to how many points lie within a small neighborhood around a point, defined by two parameters: eps (radius) and minPts (minimum neighbors). A region is considered dense if each point has at least minPts other points within distance eps, forming a cluster; sparse regions are treated as noise or separated clusters.

13. Can hierarchical clustering be used on categorical data?
    - Hierarchical clustering can be applied to categorical data, but it cannot directly use Euclidean distance and standard numeric assumptions. Instead, it relies on appropriate dissimilarity or similarity measures (e.g., Hamming, Jaccard, or Gower) or encodings of categories, then builds the hierarchy using these distances and a linkage method.

14. What does a negative Silhouette Score indicate?
    - A negative Silhouette Score for a point indicates it is, on average, closer to points in some other cluster than to points in its own assigned cluster. This suggests misassignment or overlapping/poorly separated clusters, and a low or negative average score across the dataset reflects a poor clustering structure.

15. Explain the term "linkage criteria" in hierarchical clustering.
    - Linkage criteria specify how the distance between two clusters is computed when deciding which clusters to merge next. Common choices include:

    - Single linkage: distance between the closest pair of points in the two clusters.

    - Complete linkage: distance between the farthest pair of points.

    - Average linkage: average distance between all pairs of points.

    - Ward’s linkage: increase in total within‑cluster variance if two clusters are merged.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
    - K‑Means assumes clusters are roughly spherical, of similar size, and similar density, since it optimizes a variance‑based objective around centroids. When clusters differ greatly in size or density, centroids can be pulled toward dense or large clusters, causing small or sparse clusters to be merged or split incorrectly and boundaries to be poorly placed.

17. What are the core parameters in DBSCAN, and how do they influence clustering?
    - DBSCAN has two main parameters:

    - eps (ε): the neighborhood radius around a point. A larger eps makes neighborhoods bigger, typically merging more points into clusters and reducing noise; too large can merge distinct clusters.

    - minPts: the minimum number of points required within eps for a point to be considered a core point. A larger minPts requires denser regions to form clusters, increasing robustness to noise but potentially breaking true clusters if set too high.

    - Together, eps and minPts determine what is considered a dense region, thus controlling the number, shape, and robustness of clusters and the amount of detected noise.

18. How does K-Means++ improve upon standard K-Means initialization?
    - K‑Means++ improves upon random initialization by selecting initial centroids in a more spread‑out way. It chooses the first centroid randomly, then selects each subsequent centroid with probability proportional to the squared distance from the closest already‑chosen centroid, biasing choices toward far‑away points. This reduces the chance of poor starting points, helps avoid bad local minima, and often yields faster convergence and better final clustering.

19. What is agglomerative clustering?
    - Agglomerative clustering is a bottom‑up form of hierarchical clustering. It starts with each data point as its own cluster, then iteratively merges the pair of clusters with the smallest distance (according to a chosen linkage criterion) until a stopping condition is met, such as reaching a desired number of clusters or a maximum inter‑cluster distance.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?
    - Inertia only measures how compact clusters are (sum of squared distances to centroids) and always decreases as K increases, so it does not directly assess separation between clusters and can over‑favor large K. Silhouette Score combines cohesion (how close a point is to its own cluster) and separation (how far it is from other clusters) into a bounded, interpretable [−1,1] metric, making it more informative for comparing different K or different clustering methods in terms of actual cluster quality.

```python
#23. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot.

from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

X, y = make_moons(n_samples=500, noise=0.1, random_state=42)

# Apply DBSCAN
db = DBSCAN(eps=0.2, min_samples=5)
labels = db.fit_predict(X)

# Masks for core/cluster points and outliers
core_mask = labels != -1
out_mask = labels == -1

# Plot
plt.scatter(X[core_mask, 0], X[core_mask, 1],
            c=labels[core_mask], cmap='plasma', s=20, label='Clustered points')
plt.scatter(X[out_mask, 0], X[out_mask, 1],
            c='red', s=30, label='Outliers')
plt.title('DBSCAN on make_moons (outliers in red)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Generate synthetic data
X, y = make_blobs(n_samples=500, centers=4, random_state=42)

ks = [2, 3, 4, 5]
scores = []

# Compute silhouette scores
for k in ks:
    km = KMeans(n_clusters=k, random_state=42)
    labels = km.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

# Bar chart
plt.bar(ks, scores)
plt.xlabel('Number of clusters k')
plt.ylabel('Silhouette score')
plt.title('Silhouette scores for k = 2 to 5')
plt.xticks(ks)
plt.tight_layout()
plt.show()
```

```python
#30. Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.

from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data

# Average linkage hierarchical clustering
linkage_matrix = linkage(X, method='average')

# Plot dendrogram
plt.figure(figsize=(10, 5))
dendrogram(linkage_matrix, labels=iris.target)
plt.title('Hierarchical Clustering Dendrogram (Iris, average linkage)')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.tight_layout()
plt.show()
```

```python
#31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.\

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate overlapping clusters
X, y_true = make_blobs(
    n_samples=500, centers=3,
    cluster_std=[1.2, 1.5, 1.8],
    random_state=42
)

# K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Decision boundaries
xx, yy = np.meshgrid(np.arange(-5, 8, 0.1), np.arange(-5, 8, 0.1))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=20, edgecolors='k')
plt.title('K-Means with Decision Boundaries (Overlapping Clusters)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.tight_layout()
plt.show()
```

```python
#34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load and scale data
bc = load_breast_cancer()
X = bc.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Elbow method
inertias = []
ks = range(2, 7)

for k in ks:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(ks, inertias, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method: Breast Cancer K-Means Inertia')
plt.grid(True)
plt.tight_layout()
plt.show()

print("Inertia values:", dict(zip(ks, inertias)))
```

```python
#43. Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN side-by-side.

from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt

# Generate noisy circles
X, y = make_circles(n_samples=500, factor=0.5, noise=0.08, random_state=42)

# K-Means
kmeans = KMeans(n_clusters=2, random_state=42)
labels_kmeans = kmeans.fit_predict(X)

# DBSCAN
dbscan = DBSCAN(eps=0.15, min_samples=8)
labels_dbscan = dbscan.fit_predict(X)

plt.figure(figsize=(12, 5))
plt.subplot(131)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', s=30)
plt.title('True Circles')

plt.subplot(132)
plt.scatter(X[:, 0], X[:, 1], c=labels_kmeans, cmap='coolwarm', s=30)
plt.title('K-Means (Poor separation)')

plt.subplot(133)
colors = ['red' if l == -1 else 'plasma' for l in labels_dbscan]
plt.scatter(X[:, 0], X[:, 1], c=labels_dbscan, cmap='plasma', s=30)
plt.title('DBSCAN (Good separation)')
plt.tight_layout()
plt.show()
```

```python
#44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import numpy as np

# Load Iris
iris = load_iris()
X = iris.data

# K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Silhouette analysis
sil_avg = silhouette_score(X, labels)
sil_samples = silhouette_samples(X, labels)

plt.figure(figsize=(12, 5))

plt.subplot(121)
y_lower = 10
for i in range(3):
    cluster_sil = sil_samples[labels == i]
    cluster_sil.sort()
    plt.fill_betweenx(np.arange(y_lower, y_lower + len(cluster_sil)),
                     0, cluster_sil, alpha=0.7)
    y_lower += len(cluster_sil) + 10

plt.axvline(sil_avg, color="red", linestyle="--")
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Cluster")
plt.title(f"Iris K-Means Silhouette Plot\n(Avg: {sil_avg:.3f})")

plt.subplot(122)
plt.scatter(range(len(sil_samples)), sil_samples, c=labels, s=20)
plt.axhline(sil_avg, color="red", linestyle="--")
plt.xlabel("Sample Index")
plt.ylabel("Silhouette Coefficient")
plt.title("Silhouette Coefficient per Sample")

plt.tight_layout()
plt.show()
```

```python
#47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Generate noisy blobs
X, y = make_blobs(
    n_samples=800,
    centers=3,
    cluster_std=[0.8, 1.2, 1.5],
    random_state=42
)

# Add extra noise
noise_points = np.random.uniform(-8, 8, (200, 2))
X_noisy = np.vstack([X, noise_points])

# DBSCAN
dbscan = DBSCAN(eps=0.6, min_samples=8)
labels = dbscan.fit_predict(X_noisy)

# Counts
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print(f"Number of clusters: {n_clusters}")
print(f"Number of noise points: {n_noise}")
print(f"Total points: {len(X_noisy)}")
```

```python
#48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

digits = load_digits()
X_digits = digits.data

# t-SNE to 2D
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X_digits)

# Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels = agg.fit_predict(X_tsne)

plt.figure(figsize=(12, 5))

plt.subplot(121)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=digits.target, cmap='tab10', s=15)
plt.title('Digits t-SNE (True Labels)')

plt.subplot(122)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab10', s=15)
plt.title('Agglomerative Clustering on t-SNE')

plt.tight_layout()
plt.show()
```
